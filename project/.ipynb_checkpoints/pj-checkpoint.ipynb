# extracting geographic info from photos


```python
img=gpsphoto.getGPSData('IMG_0346.JPG')
locations = (img['Latitude'],img['Longitude'])
print("\ncoordinate from the pic: " + str(locations) + "\n")
```

```python
data = data.assign(user_lat=locations[0])
data = data.assign(user_lon=locations[1])
```

```python
R = 6371 # Radius of the earth in km
lat1=data['lat']
lon1=data['lon']
lat2 = data['user_lat']
lon2 = data['user_lon']
dLat = np.deg2rad(lat2 - lat1)
dLon = np.deg2rad(lon2 - lon1)
#a = np.sin(dLat/2) * np.sin(dLat/2)+np.cos(np.deg2rad(lat1)) * np.cos(np.deg2rad(lat2)) * np.sin(dLon/2) * np.sin(dLon/2)
x =0.5 -np.cos(dLat)/2+np.cos(np.deg2rad(lat1))* np.cos(np.deg2rad(lat2)) * (1-np.cos(dLon))/2
y = 2*R * np.arcsin(np.sqrt(x))
data['distance']=y*1000
data
```

```python
possible_interest = data[data['distance'] < 1000]
possible_interest
```

```python
possible_interest = possible_interest.drop_duplicates(subset=['name'],keep='first').reset_index()
```
